```python
import pandas as pd
from pathlib import Path

# Load all cached game files into one DataFrame
raw_files = list(Path("data/raw/pbp").glob("*.parquet"))
df = pd.concat([pd.read_parquet(f) for f in raw_files], ignore_index=True)

print(f"{len(raw_files)} games | {df.shape[0]:,} rows | {df.shape[1]} cols")
df.head(20)
```
