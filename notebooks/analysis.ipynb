```python
import pandas as pd
from pathlib import Path

# Anchor to repo root by finding a known marker file (CLAUDE.md)
ROOT = next(p for p in [Path.cwd(), *Path.cwd().parents] if (p / "CLAUDE.md").exists())
pbp_dir = ROOT / "data/raw/pbp"
raw_files = list(pbp_dir.glob("*.parquet"))

print(f"CWD: {Path.cwd()}")
print(f"Looking in: {pbp_dir}  ->  {len(raw_files)} files")

# Fallback: if none found, locate parquet files anywhere under the repo
if not raw_files:
    raw_files = list(ROOT.glob("**/*.parquet"))
    print("Fallback found:", [str(f.relative_to(ROOT)) for f in raw_files[:5]])

df = pd.concat([pd.read_parquet(f) for f in raw_files], ignore_index=True)
print(f"{len(raw_files)} games | {df.shape[0]:,} rows | {df.shape[1]} cols")
df.head(20)
```
